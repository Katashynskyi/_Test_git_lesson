```python
import numpy as np
import matplotlib.pyplot as plt
import json
from google.colab import files
!wget 'https://drive.google.com/file/d/1jXMuWs00kvLsIxxflrB0lRNhpZANwQFZ/view?usp=sharing'
data=json.load(open('config_params.json'))
```

```python
##normal distribution_test_01
g,lst,count_lst=9.81,[],[]
angls=list(map(int,np.random.normal(loc=data['angls_loc'][0],scale=data['angls_scale'][0],size=data['size'][0])))#тысяча углов
vlcts=list(map(int,np.random.normal(loc=data['vlct_loc'][0],scale=data['vlct_scale'][0],size=data['size'][0])))#тысяча скоростей
for _ in range(len(angls)):#тысяча попаданий
    L=(vlcts[_]*(2*np.sin(angls[_] * np.pi / 180)*np.cos(angls[_]* np.pi / 180)))/g
    lst.append(int(L))
unic_lst=list(set(lst))
for _ in range(len(unic_lst)):#считаем сколько куда упало
    count_lst.append(lst.count(unic_lst[_]))
print('кол-во уникальных значений в normal\'e, Test_01',len(unic_lst))
plt.bar(unic_lst,count_lst)
plt.title('Как стреляет суперпушка "Дора" с normal-ом тест 1')
plt.ylabel('количество упавших снарядов')
plt.xlabel('расстояние в метрах куда достал снаряд')
plt.show()
```
